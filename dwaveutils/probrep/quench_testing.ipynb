# Testing Distribution of States Produced by Quenching

```python
import numpy as np
import pandas as pd
from dictrep import DictRep
from dwavetools import find_heff_s, make_anneal_schedule

heff_to_s = pd.read_csv("heff_to_s_DW_2000Q_2_June2018.csv")
```

### Step 1) Define 3-qubit hamiltonian and set-up problem

```python
H = {(0, 0): 'h0', (1, 1): 'h0', (2, 2): 'h0', (0, 1): 'J1', (1, 2): 'J1', (0, 2): 'J2'}
h0 = [0]; J1 = [-0.1]; J2 = [0.4];
anneal_times = np.linspace(20, 200, 1)
quench_times = [1]
hxvals = np.linspace(0, .4, 5)
hidxs = pd.Index(heff_to_s['heff'])
csv_svals = pd.Index(heff_to_s['s'])
s_to_hx = {}
svals = []
for hx in hxvals:
    s = csv_svals[hidxs.get_loc(hx, 'nearest')]
    svals.append(s)
    s_to_hx[s] = hx

schedules = [make_anneal_schedule('forward', s, ta, 0, tq)\
             for s in svals for ta in anneal_times for tq in quench_times]

problem = DictRep(H, qpu='dwave', vartype='ising', encoding='logical')
print(s_to_hx)
```

```python
for i in range(len(schedules)):
    parameters = {'h0': h0, 'J1': J1, 'J2': J2, 'anneal_schedule': [schedules[i]], 'num_reads':[1]}
    problem.populate_parameters(parameters)
    problem.call_annealer()
    problem.get_state_plot(figsize=(12, 8), filename=None, title=str(schedules[i]))
```
